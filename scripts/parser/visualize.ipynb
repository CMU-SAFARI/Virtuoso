```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('./results_reservethp.csv')
```

```python

# Display the DataFrame
df.head()

```

```python
# Create a pivot table - each row has a "Trace", each column has a "Config", and the values are the "performance_model.cycle_count" values
pivot = df.pivot(index='Trace', columns='Config', values='performance_model.cycle_count')

# Create a new 
# Now normalize the pivot table by config "radix-resthp-frag-0" (i.e., do the division "pivot['radix-resthp-frag-0'] / pivot[config_name]" for each config_name)

#For each config, do the division  pivot['radix-resthp-frag-0'] / pivot[config_name]
# normalized = pivot.div(pivot['radix-resthp-frag-0'], axis=0)

div_plot = pivot['radix-resthp-frag-0']

for col in pivot.columns:
    pivot[col] = div_plot / pivot[col]

print (pivot)


means = pivot.mean()
pivot.loc['geomean'] = means
pivot = pivot.loc[['geomean']]

config_strings = pivot.columns

# order the columns based on the last number in the string
config_strings = sorted(config_strings, key=lambda x: int(x.split('-')[-1]))

# # apply the new order to the normalized pivot table

normalized = pivot[config_strings]
print(normalized)

import matplotlib.pyplot as plt
import seaborn as sns

# o
plt.figure(figsize=(10, 10))
# Make the x-axis labels vertical
plt.xticks(rotation=90)
# Use a bar plot
sns.barplot(data=normalized, ci=None)

plt.show()
```
